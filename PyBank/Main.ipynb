```python
"""
Homework: PyBank

This script is used for analyzing the financial records of a company.

"""

# Initialize the metric variables

count = 0
total = 0
average = 0
minimum = 0
maximum = 0
change_total=0
change_count=0
min_date=''
max_date=''

# Initialize lists to hold profitable and unprofitable day profits/losses

profitable_days = []
unprofitable_days = []

# Import the pathlib and csv library

from pathlib import Path
import csv

# Set the file path

csvpath = Path('./02-Python_Homework_Instructions_PyBank_Resources_budget_data.csv')

# Initialize lists of records

dates=[]
profits=[]
monthly_change=[]

# Open the csv file as an object

with open(csvpath, 'r') as csvfile:

    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter=',')

    # Read the header row first
    csv_header = next(csvreader)

    # Read each row of data after the header
    for row in csvreader: # row -> (date, $)
        # check if total (as a variable) exists; if yes then proceed; if no then create total
        dates.append(row[0]) # (date, $)
        profits.append(row[1]) # (date, $)
        total += int(row[1]) 
        count += 1 
        
# Calculate increase/decrease in profits

for i in range(len(profits)-1): 
    monthly_change.append(int(profits[i+1])-int(profits[i]))

# Calculate total and count of increase/decrease in profits

for each_change in monthly_change: 
    change_total+=each_change
    change_count+=1
    
# Calculate the average and round to the nearest 2 decimal places

average=round(change_total/change_count, 2)

# Logic to determine minimum and maximum values

for x in monthly_change: 
    if minimum == 0:
            minimum = x
    elif x < minimum:
           minimum = x
    elif x > maximum:
            maximum = x

# a for loop to find the index of the max and min
# index is when value is the max and min
# so the date will be dates[index] (+1)
for i in range(len(monthly_change)): 
    if monthly_change[i] == maximum: 
        max_date = dates[i+1]
    if monthly_change[i] == minimum: 
        min_date=dates[i+1]

 
f = open('PyBank.txt', 'w')
print("Financial Analysis", file = f)        
print("Financial Analysis")
print("----------------------------", file = f)
print("----------------------------")
print(f"Total Months: {count}", file = f)
print(f"Total Months: {count}")
print(f"Total: {'${:,.2f}'.format(total)}", file = f)
print(f"Total: {'${:,.2f}'.format(total)}")
print(f"Average Change: {'${:,.2f}'.format(average)}", file = f)
print(f"Average Change: {'${:,.2f}'.format(average)}")
print(f"Greatest Increase in Profits: {max_date} ({'${:,.2f}'.format(maximum)})", file = f)
print(f"Greatest Increase in Profits: {max_date} ({'${:,.2f}'.format(maximum)})")
print(f"Greatest Decrease in Profits: {min_date} ({'${:,.2f}'.format(minimum)})", file =f)
print(f"Greatest Decrease in Profits: {min_date} ({'${:,.2f}'.format(minimum)})")
```
